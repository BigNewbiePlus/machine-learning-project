```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter

import gzip
import json
import time
import pickle 

# load model data from file
def load_pickle(path):
    file = open(path, 'rb')
    data = pickle.load(file)
    file.close()
    return data

# save model data to pickle file
def save_pickle(data, path):
    file = open(path, 'wb')
    pickle.dump(data, file)
    file.close()
```

```python
# 不均衡数据处理 -> 正例和负例各一半
raw_datas = []
raw_words=[]
raw_labels=[]
cate_cnt = {}

i = 0

for line in gzip.open('./out.json.gz', mode='r'):
    i = i+1
    raw = json.loads(line.decode())
    tokens = [el.strip() for el in raw['source'] if el!=' '] #去掉空格
    status = raw['status']
    splits=status.split()
    
    if len(splits)>1:
        key=splits[0]+' '+splits[1]
    else:
        key=status
        
    if key in cate_cnt:
        cate_cnt[key] = cate_cnt[key] + 1
    else:
        cate_cnt[key] = 1
    
    if key == 'Accepted':
        raw_datas.append(tokens)
        raw_words.extend(tokens)
        raw_labels.append(1)
    elif key == 'Runtime error':
        raw_datas.append(tokens)
        raw_words.extend(tokens)
        raw_labels.append(0)
    if i%20000 == 0 :
        print('process line %d'%i)  
```

```python
import collections
cnt=collections.Counter(raw_words).most_common(len(raw_words))
print('辞典个数:%d'%len(cnt))
dictionary={}
reverse_dictionary={}

for word, _ in cnt:
    idx = len(dictionary)
    dictionary[word] = idx
    reverse_dictionary[idx]=word

for data in raw_datas:
    for i in range(len(data)):
        data[i] = dictionary[data[i]]
print("datas num%d:"%(len(raw_datas)))
               
f = {}
f['datas'] = raw_datas
f['labels'] = raw_labels
               
save_pickle(f, 'codeforces_full.pkl')
print('pickle data saved')
```

```python
pos_num=0
neg_num=0
for i in range(len(raw_labels)):
    if raw_labels[i] == 1:
        pos_num=pos_num+1
    else:
        neg_num = neg_num+1
print(pos_num)
print(neg_num)
print(cate_cnt)
```

```python
# 词频分布图
total_words = len(raw_words)
inter = 500
sizes = [0 for i in range(len(cnt)//inter+1)]
labels = ['%dx'%i for i in range(len(cnt)//inter+1)]
i=0
for _,count in cnt:
    i=i+1
    sizes[i//inter] = sizes[i//inter] + count

#sizes=sizes[1:]
#labels=labels[1:]
sizes= [size*100/total_words for size in sizes]
import matplotlib.pyplot as plt

plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.show()
```
